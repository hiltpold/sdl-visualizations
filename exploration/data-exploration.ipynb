```python
import pandas as pd
import plotly.graph_objects as go
```

```python
actions ="../data/metadata-actions.csv"
data_objects ="../data/metadata-data-objects.csv"
```

```python
df_actions = pd.read_csv(actions)
df_data_objects = pd.read_csv(data_objects)
```

```python
df_actions.head(10)
```

```python
df_data_objects.head(10)
```

```python
df_actions["input_edges"] = df_actions.apply(lambda row: row["inputid"].split(","), axis=1)
df_actions["output_edges"] = df_actions.apply(lambda row: row["outputid"].split(","), axis=1)
```

```python
df_actions.head(40)
```

```python
df_input_edges = df_actions[["id","input_edges"]].explode(column="input_edges")
df_output_edges = df_actions[["id","output_edges"]].explode(column="output_edges")
```

```python
df_edges = pd.merge(df_input_edges, df_output_edges, on="id")
df_edges = df_edges.rename(columns={"input_edges": "source", "output_edges": "target"})
```

```python
df_edges["source_layer"] = df_edges["source"].map(lambda x: x.split("_")[0])
df_edges["target_layer"] = df_edges["target"].map(lambda x: x.split("_")[0])
```

```python
df_edges.head(10)
```

```python
df_top_level = df_edges[df_edges["source_layer"] != df_edges["target_layer"]].reset_index(drop=False)
```

```python
tmp1 = df_top_level[["source_layer"]].rename(columns={"source_layer": "vertex"})
tmp2 = df_top_level[["target_layer"]].rename(columns={"target_layer": "vertex"})
```

```python
vertices_all = pd.concat([tmp1, tmp2], ignore_index=True)["vertex"]
vertices_unique = vertices_all.drop_duplicates().reset_index(drop=True)
vertices_unique
```

```python
vertex_lu = { x: idx for idx,x in enumerate(vertices_unique) }
```

```python
df_top_level["source_layer_index"] = df_top_level["source_layer"].map(lambda x: vertex_lu[x])
df_top_level["target_layer_index"] = df_top_level["target_layer"].map(lambda x: vertex_lu[x])
```

```python
df_top_level
```

```python
df_top_level_grouped=df_top_level.groupby(["source_layer", "target_layer", "source_layer_index", "target_layer_index"]).count().reset_index()
```

```python
df_top_level_grouped=df_top_level_grouped[["source_layer", "target_layer", "source_layer_index", "target_layer_index","id"]].rename(columns={"id":"value"})
```

```python
df_top_level_grouped
```

```python
source_indices = df_top_level_grouped["source_layer_index"].to_list()
target_indices = df_top_level_grouped["target_layer_index"].to_list()
values = df_top_level_grouped["value"].to_list()
labels = vertices_unique.to_list()
```

```python
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = labels,
      color = "blue"
    ),
    link = dict(
      source = source_indices,
      target = target_indices,
      value = values
  ))])

fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.show()
```

```python
export_edges = df_top_level_grouped.rename(columns={"source_layer": "source", 
                                     "target_layer": "target",
                                     "source_layer_index": "source_index",
                                     "target_layer_index": "target_index"}).reset_index(drop=True)
export_edges.index.name = "index"
export_vertices = vertices_unique.to_frame(name="name")
export_vertices["group"] = export_vertices["name"].map(lambda x: x)
export_vertices.index.name = "index"
```

```python
export_edges.to_csv("../data/edges_grouped.csv")
export_vertices.to_csv("../data/vertices_grouped.csv")
```
